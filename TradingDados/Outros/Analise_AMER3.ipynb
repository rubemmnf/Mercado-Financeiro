##Informações da empresa

```python
ativo = yf.Ticker('AMER3.SA')
```

```python
#Campos confiáveis para esse ativo
#Net Income - Lucro Líquido
#Income Before Tax - Lucro antes do imposto de renda
#Selling General Administrative (SG&A) Despesas Operacionais
#Gross Profit - Lucro Bruto
#Income Tax Expense - Lucro antes do imposto de renda 
#Total Revenue - Total de Receita
#Cost Of Revenue - Custo da Receita

dre = ativo.get_financials(freq='quarterly')
dre
```

```python
dre = pd.DataFrame(dre)
dre = dre/1000
dre = dre[dre.columns[::-1]]
dre
```

##Balanço

```python
ativo.get_cash_flow(freq='quarterly')
```

```python
bal = ativo.get_balance_sheet(freq='quarterly')
bal
```

```python
#Total Assets - Total do ativo
#Total Stockholder Equity - Patrimônio Líquido

balanco = pd.DataFrame(bal)
balanco = balanco/1000
balanco = balanco[balanco.columns[::-1]]
balanco
```

```python
dre
```

```python
fig = make_subplots(rows=9,
                    cols=2,
                    row_heights=[5,5,5,5,5,5,5,5,5],
                    column_widths=[3,3],
                    subplot_titles=('Receita Bruta', 'Lucro Líquido', 'Dívida CP', 'Dívida LP',
                                    'Capital de Giro','Divida Líquida',
                                    'Margem EBITDA', 'Margem Líquida', 'ROIC', 'ROE',
                                    'Div Bruta/PL', 'CAGR 5Y', 'Preço/Lucro', 'P/VP',
                                    'EV/EBIT', 'EV/EBITDA', 'Liquidez Corrente', 'Div Yeld'),
                    shared_xaxes=False)
fig.add_trace(go.Scatter(name='Receita_bruta', x=dre.columns, y=dre.loc['OperatingRevenue'], mode='lines+markers',marker_size=15,marker_symbol='diamond'), row=1, col=1)
fig.add_trace(go.Scatter(name='Receita_liquida', x=dre.columns, y=dre.loc['NetIncome'],mode='lines+markers',marker_size=15,marker_symbol='diamond'), row=1, col=2)


fig.add_trace(go.Scatter(name='Divida_CP', x=balanco.columns, y=balanco.loc['CurrentDebt'],mode='lines+markers',marker_size=15,marker_symbol='diamond'), row=2, col=1)
fig.add_trace(go.Scatter(name='Divida_CP', x=balanco.columns, y=balanco.loc['LongTermDebt'],mode='lines+markers',marker_size=15,marker_symbol='diamond'), row=2, col=2)

fig.add_trace(go.Scatter(name='WC', x=balanco.columns, y=balanco.loc['WorkingCapital'],mode='lines+markers',marker_size=15,marker_symbol='diamond'), row=3, col=1)
fig.add_trace(go.Scatter(name='Divida_liquida', x=balanco.columns, y=balanco.loc['NetDebt'],mode='lines+markers',marker_size=15,marker_symbol='diamond'), row=3, col=2)

fig.add_trace(go.Bar(name='Margem_Ebtida', x=ind_luc.index, y=ind_luc['mrgebit']), row=4, col=1)
fig.add_trace(go.Bar(name='Margem_Liquida', x=ind_luc.index, y=ind_luc['mrgliq']), row=4, col=2)
fig.add_trace(go.Bar(name='ROIC', x=ind_luc.index, y=ind_luc['roic']), row=5, col=1)
fig.add_trace(go.Bar(name='ROE', x=ind_luc.index, y=ind_luc['roe']), row=5, col=2)

fig.add_trace(go.Bar(name='divbpatr', x=end.index, y=end['divbpatr']), row=6, col=1)
fig.add_trace(go.Bar(name='c5y', x=ind_luc.index, y=ind_luc['c5y']), row=6, col=2)

fig.add_trace(go.Bar(name='pl', x=ind_valor.index, 
                     y=ind_valor['pl']), row=7, col=1)
fig.add_trace(go.Bar(name='pvp', x=ind_valor.index, 
                     y=ind_valor['pvp']), row=7, col=2)

fig.add_trace(go.Bar(name='evebit', x=ind_valor.index, y=ind_valor['evebit']), row=8, col=1)
fig.add_trace(go.Bar(name='evebitda', x=ind_valor.index, y=ind_valor['evebitda']), row=8, col=2)

fig.add_trace(go.Bar(name='ind_liq', x=ind_liq.index, y=ind_liq['Liq. Corr.']), row=9, col=1)
fig.add_trace(go.Bar(name='dy', x=ind_luc.index, y=ind_luc['dy']), row=9, col=2)
fig.update_layout(title_text='<b>Avaliação Fundamentalista AMER3.SA<b>', 
                  template='plotly_dark',#template pré-definido da plotly
                  showlegend=False, #esconder ou mostrar legenda
                  height=1800, #altura
                  width=900) #largura
fig.show()
```
