##Indicadores de Lucratividade / Operacionais

1) Margem EBIT 

2) Margem Líquida 

3) ROIC 

4) ROE 

5) Dividend Yeld 

6) CAGR - Crescimento de receita dos últimos 5 anos

```python
df.query('index in @setor').loc[:,['mrgebit','mrgliq', 'roic', 'roe', 'dy', 'c5y']]
```

```python
ind_luc = df.query('index in @setor').loc[:,['mrgebit','mrgliq', 'roic', 'roe', 'dy', 'c5y']]

ind_luc = ind_luc.sort_values(by='roic', ascending=False)

ind_luc
```

##Indicadores de Preço & Valor

1)P/L 

2)P/VPA = Preço Ação/(PL/N ações)

3)EV/EBIT = EV - MarketCap + (div bruta - caixa)
 
4)EV/EBITDA

```python
ind_valor = df.query('index in @setor').loc[:,['pl', 'pvp', 'evebit', 'evebitda']]

ind_valor = ind_valor.sort_values(by='pl')

ind_valor
```

##Indicador de liquidez

- Liquidez_Corrente = Ativo Circulante/Passivo Circulante

```python
ind_liq = df2.query('index in @setor').loc[:,['Liq. Corr.']]

ind_liq = ind_liq.sort_values(by='Liq. Corr.', ascending=False)

ind_liq
```
