```python
# TASK 2 - fazer o import dos pacotes necessários e ja instalados

import pandas as pd
import vectorbt as vbt
import yfinance as yf
```

```python
tickers = ['ABEV3.SA', 'BBDC4.SA', 'ITUB4.SA','PETR4.SA', 'VALE3.SA', 'WEGE3.SA']
```

```python
dividendos = vbt.YFData.download(tickers,
                                 start='2012-01-01',
                                 end='2021-12-31'
                                 ).get('Dividends')

dividendos
```

Qual o total pago por ação para cada ativo nesses 10 anos avaliados (2012 a 2021)?<p>
OBS: Métrica não muito útil já que depende do preço gasto em uma ação

```python
# TASK 6 - Trazer a soma do total pago por ação e ordenação decrescente

dividendos_totais = dividendos.sum()

dividendos_totais.sort_values(ascending=False)
```

```python
# TASK 7 - Inserir uma coluna representando o "Ano" indicado pelo index usando strtime('%Y')

dividendos['Ano'] = pd.to_datetime(dividendos.index).strftime('%Y')

dividendos
```

Usar uma pivot table para visualizar os valores pagos por ano

```python
# TASK 8 - use o método pd.pivot_table passando o df de dividendos com função de agregação de soma

tab_dinamica_dividendos = dividendos.pivot_table(index='Ano',aggfunc='sum')
tab_dinamica_dividendos
```

A classificação de dividendos depende não apenas do valor nominal pago, mas do preço de cada ação naquele momento.

```python
# TASK 9 - Buscar as cotações de todos os dias. Intervalos maiores podem gerar erros e NaN. Usar a yfinance

ticker_open = yf.download(tickers, start='2012-01-01', end='2022-01-01', interval='1d')['Open']
```

```python
# TASK 10 - Inserir uma coluna representando o "Ano" indicado pelo index

ticker_open['Ano'] = pd.to_datetime(ticker_open.index).strftime('%Y')
ticker_open
```

```python
# TASK 11 - Tabela dinâmica pegando apenas o primeiro valor observado de cada ano (algo similar a remover duplicatas duplicatas)

tab_dinamica_open = ticker_open.pivot_table(index='Ano',aggfunc='first')
tab_dinamica_open
```

Agora que temos a soma dividendos e preço de abertura, usamos a fórmula do Dividend Yield (DY) para cada ano. Relembrando...

Dividend Yield (DY %) = Dividendos por ação / Preço da ação * 100

OBS: Utilizamos a soma dos dividendos e JCP entre janeiro e dezembro de cada ano por ação, dividido pelo preço de abertura da no primeiro dia de pregão do respectivo ano

```python
# TASK 12 - Como as duas tabelas estão simétricas, posso simplesmente dividir uma pela outra

tabela_DY_anual = tab_dinamica_dividendos/tab_dinamica_open
tabela_DY_anual
```

Podemos transpor o dataframe e classificar para um ano desejado

```python
# TASK 13 - Fazer o transposto da tabela de DY anual e ordenar os valores pelo ano de 2021

tabela_DY_anual = tabela_DY_anual.transpose()

tabela_DY_anual = tabela_DY_anual.sort_values(by = '2021', ascending=False)

tabela_DY_anual
```

```python
# TASK 14 - Trazer o valor acumulado dos dividendos ao longo desses anos, usar um metodo de soma cumulativa

tabela_DY_acumulado_anual = tabela_DY_anual.copy()

tabela_DY_acumulado_anual = tabela_DY_acumulado_anual.transpose().sort_index(axis=1).cumsum()*100

tabela_DY_acumulado_anual
```
